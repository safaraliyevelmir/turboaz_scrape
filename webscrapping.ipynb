```python
all_cars = []
```

```python
page_count = 1
while page_count<417:
    print(page_count)
    url = f"https://turbo.az/autos?page={page_count}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    ls = soup.find_all("a",class_='products-i__link')
    for i in ls:
        response=requests.get("https://turbo.az/"+i['href'])
        soup = BeautifulSoup(response.content, 'html.parser')
        label = soup.find_all("label",class_="product-properties__i-name")
        name = soup.find_all("span",class_="product-properties__i-value")
        dc = {}
        for i in range(len(label)):
            dc[label[i].text] = name[i].text
        all_cars.append(dc)
        
    page_count+=1
all_cars
```

```python
cars = pd.DataFrame(all_cars)
import numpy as np
```
